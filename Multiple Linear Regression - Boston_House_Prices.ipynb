### 5.3 Dealing with Outliers

```python
# Display (plot) and explore the dataset to see trends / statistics
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20,10))
index = 0
axs = axs.flatten()
for k,v in boston_df.items():
    sns.boxplot(y=k, data=boston_df, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

#### Conclusion:
- Columns like CRIM, ZN, RM, B, MEDV seems to have outliers

```python
# Percentage of Outliers in columns using InterQuartile Range (IQR) score
for k, v in boston_df.items():
        Q1 = v.quantile(0.25)   # Lower quartile
        Q3 = v.quantile(0.75)   # Upper quartile
        IQR = Q3 - Q1           # IQR score
        
        cut_off = 1.5 * IQR     # Calculate the Outlier cutoff
        lower = Q1 - cut_off
        upper = Q3 + cut_off
        
        v_col = v[(v <= lower) | (v >= upper)]                       # Detecting Outliers with IQR score
        perc = np.shape(v_col)[0] * 100.0 / np.shape(boston_df)[0]   # Percentage of Outliers in columns
        print("Column %s outliers = %.2f%%" % (k, perc))
```

```python
# Shape of the datatset (before removing Outiliers)
boston_df.shape
```

```python
# Remove Outliers in the target variable (label)
boston_df = boston_df[~(boston_df['MEDV'] >= 50.0)]
```

```python
# Remove Outliers from all columns in the dataset
boston_df = boston_df[~((boston_df['MEDV'] < (Q1 - 1.5 * IQR)) | (boston_df['MEDV'] > (Q3 + 1.5 * IQR)))]
```

```python
# Shape of the datatset (after removing Outiliers)
boston_df.shape
```

#### Conclusion:

- 40 out of 506 values (or 7.91% of total number of the values) in the target variable (label) 'MEDV' are Outliers
- Outliers in the target variable (label) 'MEDV' are removed using IQR
    - IQR is a measure of statistical dispersion being equal to the difference between 75% (Q3) and 25% (Q1), or between upper and lower quartiles

### 6.7 Evaluate the model on Testing data

#### Linear Regression Evaluation metrics

- The goal is to create the best fitted model i.e. to minimize the Linear Regression metrics (errors) (minimum cost)		
- Evaluation of the model:
        1. Determine the best fitted Linear Regression equation (line) that will be used to make predictions
        2. Minimize the Loss (Cost) Function	
- Loss (Cost) Function is the difference between actual output value (Yi) and predicted output value (Ŷi)	
- Loss (Cost) Functions for Linear Regression are:	

    - **Mean Absolute Error (MAE)**
    $$\frac 1n\sum_{i=1}^n|y_i-\hat{y}_i|$$
    
    - **Mean Squared Error (MSE)**
    $$\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2$$
    
    - **Root Mean Squared Error (RMSE)**
    $$\sqrt{\frac 1n\sum_{i=1}^n(y_i-\hat{y}_i)^2}$$

```python
# Import metrics function from Scikit-Learn
from sklearn import metrics
```

```python
# MAE - Mean of the absolute value of the errors
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))
```

```python
# MSE - Mean of the squared errors
print('MSE:', metrics.mean_squared_error(y_test, y_pred))
```

```python
# RMSE - Square root of the mean of the squared errors
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

##### Conclusion:
- As the RMSE value < 5, the ability of the model to accurately predict the data is good
